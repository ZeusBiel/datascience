Importações dos modulos

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

Lendo o arquivo csv e armazenando na variavel base

```python
#lendo arquivo csv
base = pd.read_csv('onlinefoods.csv')
#Mostrando dados do arquivo
display(base)
```

Usando drop para retirar as colunas indejadas

```python
#Retirando as colunas "Unnamed: 12" e "Pin code" da variavel base
base = base.drop(['Unnamed: 12','Pin code'],axis=1)
display(base)
```

```python
#Fazendo a dumificação dos dados de cada coluna necessária, assim todos os dados podem ser convertidos para float sem erro
base_categoricos = pd.get_dummies(base,columns=["Gender","Marital Status","Occupation","Monthly Income","Educational Qualifications","Output"], dtype=int)
display(base_categoricos)
```

Declarando as variaveis X e para treinamento

```python
#Declarando valores de "Feedback" para y, que é o alvo
y = base_categoricos['Feedback']
#Declarando todos o dados exceto "Feedback" para x, que são os dados para prever o alvo
X = base_categoricos.drop(['Feedback'],axis=1)

display(y)
display(X)
```

Treinamento e teste do modelo

```python
#Treinando o modelo com as variaveis
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
#Declarando a árvore e decisão, que tenta prever qual será o resultado alvo de acordo com os dados estudados
dec = DecisionTreeClassifier()
#Testando o modelo
dec.fit(X_train, y_train)
#Testando a acurácia do modelo
score = dec.score(X_test, y_test)
display(score)
```
